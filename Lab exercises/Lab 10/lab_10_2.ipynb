```python
import matplotlib.pyplot as plot
```

#### Line graph

```python
x = [1, 2, 3, 4, 5]
y = [1, 4, 9, 16, 25]
```

```python
plot.plot(x, y, color="blue")
plot.xlabel("Integers")
plot.ylabel("Squares")
plot.title("y = x**2")
plot.show()
```

#### Bar graph

```python
courses = ["CP", "BEE", "FENG", "PST", "CALC"]
marks = [19, 20, 18, 20, 18]
```

```python
plot.bar(courses, marks, color="green")
plot.xlabel("Courses")
plot.ylabel("Marks")
plot.title("Marks against Courses")
plot.show()
```

#### Scatter graph

```python
a = [5,7,8,7,2,17,2,9,4,11,12,9,6]
b = [99,86,87,88,111,86,103,87,94,78,77,85,86]
```

```python
plot.scatter(a, b)
plot.show()
```
