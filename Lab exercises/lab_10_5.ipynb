```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
arr1 = np.array([1, 2, 3, 4, 5])
print(arr1)
```

```python
arr2 = np.array([[1, 2], [3, 4], [5, 6]])
print(arr2)
```

```python
a = np.array([0, np.pi/2, np.pi, 3*np.pi/2])
print(a)
```

```python
x = np.linspace(-np.pi, np.pi, 201)
sin = np.sin(x)
cos = np.cos(a)
tan = np.tan(a)
print(sin)
```

```python
plt.plot(x, sin, label="sin")
plt.plot(a, cos, label="cos")
plt.plot(a, tan, label="tan")
plt.legend()

plt.xlabel("Angle (rad)")
plt.ylabel("Y-axis")
plt.title("Trigonometric Graphs")

plt.show()
```
